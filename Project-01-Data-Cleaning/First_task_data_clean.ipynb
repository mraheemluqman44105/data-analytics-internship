# Data Cleaning Project
This project includes data preprocessing, handling missing values, and basic analysis using Pandas.

## 1. Import Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```


## 2. Load Dataset


```python
data = pd.read_excel('Dataset for Data Analytics.xlsx')
data
```


## 3. Data Overview



```python
# Check data types and missing values
data.info()
```

```python
data = data.drop_duplicates()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.describe()
```


## 4. Data Cleaning


```python
data = data.drop_duplicates()
```

```python
data.isnull()
```

```python
data['CouponCode'] = data['CouponCode'].fillna('NO_COUPON')
```

```python
data.head()
```

```python
data['UnitPrice'] = data['UnitPrice'].round(2)
data['TotalPrice'] = data['TotalPrice'].round(2)
```

## 5. Data Visualization

```python
sns.histplot(data['UnitPrice'],kde = True , bins = 8)
plt.show()
```

```python
sns.histplot(data['TotalPrice'],kde = True )
plt.show()
```

```python
sns.countplot(x = 'PaymentMethod' , data = data)
plt.show()
```

```python
data['OrderStatus'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Order Status Distribution")
plt.show()
```

```python
sns.scatterplot(x= 'UnitPrice', y='TotalPrice',data=data , hue = 'Product')
plt.title("Price Relationship")
plt.grid()
plt.show()
```

```python
sns.countplot(x = 'CouponCode' , data =data)
plt.show()
```

```python
data.to_excel('Cleaned_Dataset.xlsx',index = False)
```
